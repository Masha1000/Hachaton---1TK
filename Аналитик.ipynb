```python
list = []
for i in result:
    shedict = {"наименование":[], "номер":[], "год":[], "завод":[], "комментарий":[]}
    #print(i)
    
    text = i.split(' ')
    for i in range(len(text)):   #text -> список строк из расписания.
        #print(text)
        #print(i)
        #print(text[i])
        if text[i] in shedict:
           #print(i)
            key = text[i]
            continue
        else:
            shedict[key].append(text[i])
    list.append(shedict)
#print(list)            
#print(*shedict['номер'], sep='\n') 
```
